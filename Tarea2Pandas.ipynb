```python
#Ejercicio 1
import pandas as pd

df = pd.read_csv('/content/ventas.csv')
df
```

```python
#Calcule la cantidad total de productos vendidos por categoría.
df.groupby('Producto')['Cantidad'].sum()

# Determine cuál es el producto con el mayor total de ventas.
df.groupby('Producto')['Cantidad'].sum().idxmax()

# Encuentre el precio promedio de los productos vendidos.
df.groupby('Producto')['Precio_Unitario'].mean().rename('Precio_Promedio')
```

```python
#Ejercicio2 Análisis del Clima
df_clima = pd.read_csv('/content/clima.csv')
df_clima
```

```python
#Determine el registro con la temperatura más alta y el más bajo en el dataset
df_clima.groupby('Ciudad')['Temperatura'].mean().rename('Temperatura Promedio')

#Identifique qué ciudad tuvo la temperatura más alta y cuál la más baja en todo el dataset.
df_clima[df_clima['Temperatura'] == df_clima['Temperatura'].max()]
df_clima[df_clima['Temperatura'] == df_clima['Temperatura'].min()]

#Encuentre cuántos registros tienen una temperatura mayor a 30°C.
df_clima.loc[df_clima['Temperatura'] > 30, :] #.count -> Resultado: 50

# Cuenta cuántos días en total hay registrados por cada ciudad.
df_clima.groupby('Ciudad')['Fecha'].count()
```

```python
#Ejercicio 3: Estadísticas de Calificaciones
df_calificaciones = pd.read_csv('/content/calificaciones.csv')
df_calificaciones
```

```python
#Cargue el dataset y muestre las primeras filas.
df_calificaciones.head()

# . Calcule el promedio de calificaciones por materia.
df_calificaciones.groupby('Materia')['Calificación'].mean()

# Identifique el estudiante con el promedio más alto
df_promedio = df_calificaciones.groupby('Estudiante')['Calificación'].mean().rename('Promedio')
df_promedio[df_promedio == df_promedio.max()]

#. Agrupa las calificaciones por estudiante y calcule el promedio de cada uno.
df_calificaciones.groupby('Estudiante')['Calificación'].mean()

# . Identifique cuántos estudiantes tienen un promedio superior a 85.
df_promedio[df_promedio > 85]

#Encuentre la materia con la mayor cantidad de calificaciones registradas.
df_calificaciones.groupby('Materia')['Calificación'].count().idxmax()

#Muestre los 5 estudiantes con el promedio más bajo.
df_promedio.sort_values(ascending=True).head()
```
